# 색상 추가하는 방법

```python
from palettable.colorbrewer.qualitative import Dark2_3 as pal
import random
```

```python
pal.colors
```

```python
def color_func(**kwargs):
    return tuple(pal.colors[random.randint(0,len(pal.colors)-1)])
```

```python
wc.recolor(color_func=color_func)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(wc)
plt.axis('off')
```

# 예쁘게 만들어보기

```python
#배달의 민족 주아체 : BMJUA_ttf.ttf
wc=WordCloud(font_path="c:/Windows/Fonts/malgun.ttf",
             background_color="#ffffff",
             max_words=3000,
             mask=aladdin_mask,
             stopwords=sw,
             random_state=7)
wc=wc.generate(text)
```

```python
def color_f(**kwargs):
    list = [[236,64,122],[156,39,176],[239,83,80],[103,58,183],[224,64,251],[66,165,245],[41,182,246],[24,255,255]]
    #3, EF5350
    #673AB7
    #E040FB
    #42A5F5
    #29B6F6
    #18FFFF
    return tuple(list[random.randint(0,len(list)-1)])
```

```python
wc.recolor(color_func=color_f)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(wc)
plt.axis('off')
```
